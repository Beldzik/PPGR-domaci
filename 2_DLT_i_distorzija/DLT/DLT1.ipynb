```python
import numpy as np
import numpy.linalg as la

O3=np.array([0, 0, 0])

def lema_matrica(P1, P2):
    prva=np.concatenate((O3, -P2[2]*P1, P2[1]*P1))
    druga=np.concatenate((P2[2]*P1, O3, -P2[0]*P1))
    return np.array([prva, druga])

def DLT(PS1, PS2):
    A=[]
    for i in range(len(PS1)):
        M=lema_matrica(PS1[i], PS2[i])
        A.append(M[0])
        A.append(M[1])
    A=np.array(A)
    U, S, Vh = la.svd(A)
    P=Vh[8].reshape(3, 3)
    return P/P[2, 2]
    

M1=np.array([1,2,3])
M2=np.array([3,2,1])
M3=np.array([1, -2, 1])
M4=np.array([1,1,1])
M5=np.array([0,-2,3])
M6=np.array([1,0,-1])

n=8
M1p=np.array([21.3 + n, -3.2, 7.7 + 0.4*n])
M2p=np.array([11.1 + 3*n, 1.6, 15.1 + 0.4*n])
M3p=np.array([-0.8 + n, -6.5, 5.7 - 0.4*n])
M4p=np.array([8.1 + n, -0.4, 5.8 + 0.2*n])
M5p=np.array([9.3, -11.2, 3. - 0.4*n])
M6p=np.array([-5.1 + n, 2.4, 3.7])

PS1=[M1, M2, M3, M4, M5, M6]
PS2=[M1p, M2p, M3p, M4p, M5p, M6p]
rez=DLT(PS1, PS2)
print(rez)
```
